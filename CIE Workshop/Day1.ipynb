```python
# imports
import numpy as np
import pandas as pd
```

```python
step_data = [3620, 7891, 9761, 3907, 4338, 5373]
step_counts = pd.Series(step_data, name='steps')
print(step_counts)
```

```python
step_counts.index = pd.date_range('20150329',periods=6)
step_counts
```

```python
step_counts['2015-04']
```

## Functions:
- dtypes

```python
step_counts = step_counts.astype(np.float64)
step_counts
```

```python
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9, 0, None]
joined_data = list(zip(step_data, cycling_data))
print(joined_data)
activity_df = pd.DataFrame(joined_data)
activity_df
```

```python
activity_data = pd.DataFrame(activity_df, index=pd.date_range('20150329',periods=6),columns=['cycling','walking'])
activity_data
```

```python
#OOPS approach
activity_data.walking
```
