```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [2,4,5,6,9]
s = pd.Series(data)
s
```

```python
s.index=pd.date_range('20230605', periods=5)
s
```

```python
# s = s.apply(lambda x: x / 2)
print(s.apply(lambda x: x / 2))
```

```python
# data2 = [1,2,3,4,5]
final_data = list(zip(s, s.apply(lambda x: x / 2)))
print(final_data)
df = pd.DataFrame(final_data)
df
```

```python
df['c3'] = [12,14,15,16,19]
df.columns=['c1','c2','c3']
df
```

```python
plt.plot(df['c1'], df['c3'],ls="-.",marker="o")
```

```python
df.columns = [x.replace("c", f"column") for x in df.columns]
```

```python
df
```
