```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# inline is a magic function that allows us to see the plot in the notebook
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.title("Random Data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
from matplotlib import style as style
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]

style.use('ggplot')
plt.plot(x,y, label="First Line",linewidth=3)
plt.plot(x2,y2, label="Second Line")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Random Data")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, label="First Line",linewidth=3)
plt.scatter(x2,y2, label="Second Line")
plt.legend()
```

```python
# bins is the number of bars
plt.hist(np.random.randn(10000), bins=50)
plt.show()
```
