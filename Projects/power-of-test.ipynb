```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
n = 30
# H0 : mu >= 50
# H1 : mu < 50 // left tailed test, Power : area to the left of the critical value
mu0 = 50
sigma = 10
alpha = 0.05
z_nd = norm.ppf(0.05)
se = sigma/np.sqrt(n)
critical_value = z_nd*se + mu0
power = []
# area to left is given by cdf
for i in range(40, mu0):
    z_ad = (critical_value - i)/se
    power.append(norm.cdf(z_ad))
plt.plot(range(40, mu0), power)
plt.xlabel('mu')
plt.ylabel('power')
plt.show()
```

```python
n_list = [10,30,100,500]
z_nd = norm.ppf(0.05) #alpha = 0.05
mu0 = 50
sigma = 10
alpha = 0.05
power = []
for sample_size in n_list:
    se = sigma/np.sqrt(sample_size)
    critical_value = z_nd*se + mu0
    power = []
    for i in range(40, mu0):
        z_ad = (critical_value - i)/se
        power.append(norm.cdf(z_ad))
    label_value = 'n = {}'.format(sample_size)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()        
```

```python
# for different sigma
n = 30
z_nd = norm.ppf(0.05) #alpha = 0.05
mu0 = 50
sigma_list = [5,10,20]
alpha = 0.05
power = []
for sigma in sigma_list:
    se = sigma/np.sqrt(n)
    critical_value = z_nd*se + mu0
    power = []
    for i in range(40, mu0):
        z_ad = (critical_value - i)/se
        power.append(norm.cdf(z_ad))
    label_value = 'sigma = {}'.format(sigma)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()
```

```python
# for different alpha
n = 30
z_nd = norm.ppf(0.05) #alpha = 0.05
mu0 = 50
sigma = 10
alpha_list = [0.01,0.05,0.1]
power = []
for alpha in alpha_list:
    z_nd = norm.ppf(alpha)
    se = sigma/np.sqrt(n)
    critical_value = z_nd*se + mu0
    power = []
    for i in range(40, mu0):
        z_ad = (critical_value - i)/se
        power.append(norm.cdf(z_ad))
    label_value = 'alpha = {}'.format(alpha)
    plt.plot(range(40, mu0), power, label=label_value)
    plt.legend(loc = "lower left")
plt.show()
```
